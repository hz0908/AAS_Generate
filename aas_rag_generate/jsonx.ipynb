```python
import json


def parse_json(json_text):
    # 使用 json.loads 方法将 JSON 字符串解析为 Python 对象
    data = json.loads(json_text)
    equipment_info = []
    for item in data["equipmentInfo"]:
        equipment = {
            "equipmentName": item["equipmentName"],
            "quantity": item["quantity"],
            "instanceEquipment": []
        }
        for instance in item["instanceEquipment"]:
            instance_equipment = {
                "descriptionName": instance["descriptionName"],
                "programName": instance["programName"],
                "position": instance["position"]
            }
            equipment["instanceEquipment"].append(instance_equipment)
        equipment_info.append(equipment)
    return equipment_info


# 假设 json_text 是存储上述 JSON 数据的字符串
json_text = '''{
"equipmentInfo": [
{
"equipmentName": "franka 机械臂",
"quantity": 2,
"instanceEquipment": [
{
"descriptionName": "franka 机械臂 A",
"programName": "franka_arm_A",
"position": "0,0,0"
},
{
"descriptionName": "franka 机械臂 B",
"programName": "franka_arm_B",
"position": "0,0,0"
}
]
},
{
"equipmentName": "ur10 机械臂",
"quantity": 1,
"instanceEquipment": [
{
"descriptionName": "ur10 机械臂",
"programName": "ur10_arm",
"position": "0,0,0"
}
]
},
{
"equipmentName": "工作台",
"quantity": 1,
"instanceEquipment": [
{
"descriptionName": "工作台",
"programName": "workbench",
"position": "0,0,0"
}
]
},
{
"equipmentName": "托盘",
"quantity": 1,
"instanceEquipment": [
{
"descriptionName": "托盘",
"programName": "pallet",
"position": "0,0,0"
}
]
},
{
"equipmentName": "运输车",
"quantity": 1,
"instanceEquipment": [
{
"descriptionName": "运输车",
"programName": "transport_vehicle",
"position": "0,0,0"
}
]
}
]
}'''

result = parse_json(json_text)
print(result)
```
